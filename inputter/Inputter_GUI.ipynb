```python
from ttkwidgets.autocomplete import AutocompleteCombobox
from tkinter import *

def get_units(surka):
    units_dict = {}
    with open("../OCR_test/universum.surky") as file:
        lines = file.readlines()

    for line in lines:
        units_dict.update({line.split(',')[0].strip(): line.split(',')[1].strip()})

    return units_dict[surka]

def add_row():
    global row_index

    # Create widgets for the new row
    minus_button = Button(frame, text='-', font=('Arial', 18), command=lambda row=row_index: remove_row(row))
    minus_button.grid(row=row_index, column=0, padx=(5, 0), pady=5)

    label_text = f'surovina {row_index}:'
    label = Label(frame, text=label_text, font=('Arial', 14), bg='#FFFFFF')
    label.grid(row=row_index, column=1, padx=10, pady=5, sticky='w')

    entry = AutocompleteCombobox(frame, width=30, font=('Arial', 18), completevalues=surky_universe)
    entry.grid(row=row_index, column=2, padx=10)
    entry.bind("<<ComboboxSelected>>", lambda event, row=row_index: update_label(row))
    entry.bind("<KeyRelease>", lambda event, row=row_index: update_label(row))
    entry.bind("<KeyRelease>", lambda event, row=row_index: autocomplete(row, event))
    entry_fields.append(entry)

    kolik_entry = Entry(frame, font=('Arial', 18))
    kolik_entry.grid(row=row_index, column=3, padx=10)
    kolik_entries.append(kolik_entry)

    label_text_var = StringVar()
    label_text_var.set("<jednotky>")
    label_texts.append(label_text_var)
    new_label = Label(frame, textvariable=label_text_var, font=('Arial', 14), bg='#FFFFFF')
    new_label.grid(row=row_index, column=4, padx=10, sticky='w')

    # Increment row index for the next row
    row_index += 1

def remove_row(row):
    # Remove the widgets associated with the specified row
    for widget in frame.grid_slaves():
        if int(widget.grid_info()["row"]) == row:
            widget.grid_forget()

def write_fields():
    recipe = recipe_entry.get("1.0", "end-1c")
    print("Recipe:", recipe)

    for i, widget in enumerate(entry_fields):
        if widget.get() != "":
            print(f"{label_texts[i].get()}: {widget.get()}, Kolik: {kolik_entries[i].get()}")

# Initialize row index
row_index = 0
entry_fields = []
kolik_entries = []
label_texts = []

with open("../OCR_test/universum.surky") as file:
    surky_file_lines = file.readlines()

surky_universe = [line.split(",")[0] for line in surky_file_lines]

def update_label(row):
    selected_word = entry_fields[row].get()
    if selected_word in surky_universe:
        label_texts[row].set(get_units(selected_word))
    else:
        label_texts[row].set("<jednotky>")

def autocomplete(row, event):
    current_text = entry_fields[row].get()
    matching_options = [option for option in surky_universe if option.startswith(current_text)]

    if matching_options:
        entry_fields[row].set_completion_list(matching_options)

def quit_app(event):
    ws.destroy()

ws = Tk()
ws.title('Surky Explorer')
ws.geometry('1000x400')
ws.config(bg='#FFFFFF')

# Create a label for "Surky" above all rows
Label(ws, bg='#FFFFFF', font=('Arial', 18), text='Surky:').grid(row=0, column=0, sticky='w')

frame = Frame(ws, bg='#FFFFFF')
frame.grid(row=1, column=0)

# Call add_row() once at the start to ensure the first row has a label
add_row()

# Bind the 'Escape' key to quit the application
ws.bind("<Escape>", quit_app)

# Add button to add a new row
add_button = Button(ws, text='+', font=('Arial', 18), command=add_row)
add_button.grid(row=2, column=0, pady=10)

# Add text-box for typing recipe
recipe_label = Label(ws, bg='#FFFFFF', font=('Arial', 18), text='Recept:')
recipe_label.grid(row=row_index + 2, column=0, sticky='w')

recipe_entry = Text(ws, font=('Arial', 14), height=4, width=50)
recipe_entry.grid(row=row_index + 3, column=0, padx=10, pady=5, sticky='w')

# Add button to write non-empty fields
write_button = Button(ws, text='Pridat recept', font=('Arial', 18), command=write_fields)
write_button.grid(row=row_index + 4, column=0, pady=10)

ws.mainloop()
```
